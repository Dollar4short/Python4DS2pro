## Class Work

clean the dataset titanic_train.csv in the file folder

Load the dataset using pandas framework
check for missing values
replace the missing values with mean or mode
and some visualisations will be nice if added 


```python
df2=pd.read_csv('Downloads/titanic_train.csv')
df2
```

```python
df2.info()
```

from the results above we can see that the following columns [age,fare,cabin,embarked,boat,home.dest,body]
have missing values lets examine the percentages to know whom we shall fill and those we would remove 

```python
df2.isnull().sum()* 100/len(df2)
```

we can see that cabin, boat, body and home.dest have a high percentage of missing values we shall remove them from our dataset and fill the age, fare and embarked respectively with thier mean

```python
df2.columns
```

```python
to_drop2=['cabin','boat', 'body', 'home.dest']

#dropping columns with high percentages
df2.drop(to_drop2, inplace=True, axis=1)

```

```python
df2.info()
```

Filling age, fare with their means respectively and embarked with mode because it is filled with alphabets/categorical

```python
# Fill NA values with mean for age and fare, mode for embarked
df2_filled = df2.fillna(df2[['age', 'fare']].mean())  # Fill age and fare with mean 
df2_filled = df2_filled.fillna(df2['embarked'].mode().iloc[0])  # Fill embarked with mode

print(df2_filled)
```

```python
df2_filled.info()
```

```python
#converting age to integers to avoid decimal points in the age 
df2_filled['age']=df2_filled['age'].astype(int)
```

```python
df2_filled.head()
```

Having fixed that we examined the data types and thought it best to convert the sex column to integers and map them as 0s and 1s

```python
# Create a dictionary to map text labels to numeric 
sex_map = {'female': 0, 'male': 1}

# Apply the mapping to the sex column
df2_filled['sex'] = df2_filled['sex'].replace(sex_map)

df2_filled
```

```python
#Doing some visualisations
import seaborn as sns

correlation = df2_filled.corr()

plt.figure(figsize=(10, 6))

sns.heatmap(correlation, annot=True, cmap='coolwarm')

plt.xlabel('Features')
plt.ylabel('Features')
plt.title('Correlation Matrix')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar chart for 'embarked' and 'Survived'
sns.barplot(x='embarked', y='survived', data=df2_filled)  # Use seaborn for bar chart

plt.xlabel('Embarked')
plt.ylabel('Survival Rate')
plt.title('Survival Rate vs Embarked')

plt.show()
```
